# Energy Consumption Forecast


## Problem Statement
Forecast the electricity consumption of top the household with highest number of samples on an hourly basis based on the previous usage pattern.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import tensorflow as tf
import seaborn as sns
from datetime import date
import matplotlib.dates as mdates
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from statsmodels.tsa.arima_model import ARIMA
plt.style.use('fivethirtyeight')
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

f'TF Version: {tf.__version__}'
```

```python
df = pd.read_csv('Power-Networks-LCL.csv')
df.head()
```

## Preprocessing

```python
# Checking for null values

f'Null Values : {df.isna().values.astype("int").sum()}'
```

```python
# Fixing Time Index

pd_dtime = pd.to_datetime(df['DateTime'])
df['Date'] = pd_dtime.dt.date
df['Time'] = pd_dtime.dt.time
```

```python
# Dropping unnecessary columns

df.drop(['DateTime', 'stdorToU', 'Acorn', 'Acorn_grouped'], axis=1, inplace=True)
```

```python
df.head()
```

## Exploring the top household usage

```python
top = df.groupby('LCLid').count().sort_values(by=['KWh'], ascending=False)[:1]
top_house_id = top.index.values.item()
top_house_id
```

The house with the ID _MAC000018_ is the household with the highest number of samples

```python
top_usage = df[df['LCLid'] == top_house_id]
top_usage.head()
```

```python
def draw_plot(x, y, title):
    fig, ax = plt.subplots()
    plt.plot(x, y)
    plt.title(title)
    plt.xlabel('Time')
    plt.ylabel('KWh of Energy')
    fig.autofmt_xdate()
    plt.show() 
```

```python
# Overall Distribution

draw_plot(x=top_usage['Date'], y=top_usage['KWh'],
          title='Overall Data Distribution of KWh variable')
```

```python
# Yealy Distribution

def draw_year_plot(year):
    date_1 = date(int(year), 1, 1)
    date_2 = date(int(year) + 1, 1, 1)

    mask = (top_usage['Date'] >= date_1) & (top_usage['Date'] < date_2)
    year_data = top_usage[mask]
    draw_plot(x=year_data['Date'], y=year_data['KWh'],
              title=f'Consumption in {year}')
```

```python
[draw_year_plot(year) for year in ['2012', '2013', '2014']]
```

```python
# Daily Consumption

def draw_date_plot(date):

    mask = top_usage['Date'] == date
    date_data = top_usage[mask]
    
    dates = map(lambda date_ : date_.strftime("%H"), date_data['Time'].values)
    
    draw_plot(x=list(dates), y=date_data['KWh'],
              title=f'Consumption on {date.strftime("%d-%m-%y")}')
```

```python
draw_date_plot(date(2013, 5, 24))
```

```python
# Monthly averages

def monthly_plot(year):
    months = ['Jan', 'Feb', 'Mar', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    month_avgs = []
    for i, month in enumerate(months, start=1):
        
        date_1 = date(int(year), i, 1)
        date_2 = date(int(year) + 1, i, 1)

        mask = (top_usage['Date'] >= date_1) & (top_usage['Date'] < date_2)
        month_data = top_usage[mask]
        month_avgs.append(month_data['KWh'].values.mean())
    
    plt.bar(months, month_avgs)
    plt.plot(month_avgs, c='orange')
    plt.title(f'Monthly averages in {year}')
    plt.xlabel('Months')
    plt.ylabel('Average Kwh')
    plt.show()
```

```python
[monthly_plot(year) for year in ['2012', '2013', '2014']]
```

## Data Preparation

 The problem statement has specified that **hourly** analysis has to be performed.Hence the **timestep** has to be **2** as each data point is recorded for each half hour.

```python
X, Y = [], []

for i in range(0,top_usage['KWh'].count() - 2):
    X.append([top_usage['KWh'].values[i],top_usage['KWh'].values[i+1]])
    Y.append(top_usage['KWh'].values[i+2])

X, Y = np.array(X), np.array(Y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.001, random_state=42, shuffle = True)
```

```python
print('X_train Shape: ', X_train.shape)
print('X_testn Shape: ', X_test.shape)
print('Y_train Shape: ', y_train.shape)
print('Y_test Shape: ', y_test.shape)
```

```python
reg = LinearRegression()
reg.fit(X_train, y_train.reshape(-1,1))
```

```python
y_pred = reg.predict(X_test)
```

```python
plt.plot(y_pred,c='orange', label = 'predicted values')
plt.plot(y_test,c='blue', label = 'actual values')
plt.legend()
plt.show()
```

```python
 mean_absolute_error(y_test, y_pred)
```

## ARIMA

```python
model = ARIMA(top_usage['KWh'].values[:X_train.shape[0]], order=(5,1,0))
model_fit = model.fit(disp=0)
print(model_fit.summary())
```

```python
y_pred_arima = [model_fit.forecast()[0] for _ in y_test]
```

```python
plt.plot(y_pred_arima,c='orange', label = 'predicted values')
plt.plot(y_test.flatten(),c='blue', label = 'actual values')
plt.legend()
plt.show()
```

## LSTM

```python
epochs = 12

batch_size = 32
```

```python
totl_len = X_train.shape[0]
max_len = totl_len - (totl_len % batch_size)
max_len
```

```python
X_train = X_train[:max_len]
y_train = y_train[:max_len]
```

```python
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
```

## Building the model and Training


```python
# Building the model
regressor_mae = tf.keras.Sequential()
regressor_mae.add(tf.keras.layers.CuDNNLSTM(100, stateful=True, return_sequences=True, input_shape=(2, 1), batch_size=batch_size))
regressor_mae.add(tf.keras.layers.CuDNNLSTM(100, stateful=True, return_sequences=True))
regressor_mae.add(tf.keras.layers.CuDNNLSTM(100, stateful=True, return_sequences=True))
regressor_mae.add(tf.keras.layers.Flatten())
regressor_mae.add(tf.keras.layers.Dense(1))
regressor_mae.compile(optimizer='adam', loss = 'mae')
regressor_mae.summary()
```

```python
for i in range(epochs):
    print("Epoch: " + str(i))
    regressor_mae.fit(X_train,  y_train, shuffle=False, epochs = 1)
    regressor_mae.reset_states()
```

```python
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], 1))[:batch_size]
y_pred_lstm = regressor_mae.predict(X_test.reshape((X_test.shape[0], X_test.shape[1], 1)), batch_size=batch_size)
```

```python
plt.plot(y_pred_lstm,c='orange', label = 'predicted values')
plt.plot(y_test.flatten(),c='blue', label = 'actual values')
plt.legend()
plt.show()
```

```python
mean_absolute_error(y_test[:batch_size], y_pred_lstm)
```
